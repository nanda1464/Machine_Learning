```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
```

```python
# 1. Baca file CSV
df = pd.read_csv("dataset_harga_rumah.csv")

# 2. Tampilkan kolom untuk identifikasi fitur yang bisa dipakai
print("Kolom-kolom dataset:")
print(df.columns, "\n")

# 3. Ambil 200 data dan pilih fitur serta target
# Misalnya kita gunakan GrLivArea (luas rumah) untuk memprediksi SalePrice (harga jual)
data = df[['Luas (m²)', 'Harga (Rp)']].dropna().head(200)
X = data[['Luas (m²)']]
y = data['Harga (Rp)']

# 4. Buat model dan latih
model = LinearRegression()
model.fit(X, y)

# 5. Prediksi dan evaluasi
y_pred = model.predict(X)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

# 6. Tampilkan hasil training dan evaluasi
print("Sampel Data:")
print(data.head(), "\n")

print("Hasil Training Model:")
print(f"Intercept (b0): {model.intercept_:.2f}")
print(f"Koefisien (b1): {model.coef_[0]:.2f}\n")

print("Evaluasi Model:")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared Score (R²): {r2:.2f}")

# 7. Visualisasi
plt.figure(figsize=(10,6))
plt.scatter(X, y, color='blue', label='Data Asli')
plt.plot(X, y_pred, color='red', linewidth=2, label='Garis Regresi')
plt.title('Prediksi Harga Rumah Berdasarkan Luas Bangunan')
plt.xlabel('Luas (m²) (Luas Rumah)')
plt.ylabel('Harga (Rp) (Harga Jual)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
